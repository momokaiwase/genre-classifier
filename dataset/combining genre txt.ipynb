```python
import os
import pandas as pd
```

```python
###COUNTRY DF
```

```python
country_txt_direct = r"C:\Users\PEREZRT\Documents\Country"

country_songs = []
```

```python
# Loop through all .txt files in the directory

for songname in os.listdir(country_txt_direct):
    if songname.endswith('.txt'):  # Process only .txt files
        file_path = os.path.join(country_txt_direct, songname)
        
        # Open the file and read its lines
        with open(file_path, 'r') as file:
            lines = file.readlines()  # Read all lines in the file
            
            # For each line, add it as a new row in the list, with the file name as additional info
            for line in lines:
               country_songs.append({'text': line.strip(), 'filename': songname})
```

```python
countrydf = pd.DataFrame(country_songs)
```

```python
countrydf
```
