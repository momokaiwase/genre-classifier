```python
#removing stopwords from songs420
from nltk.corpus import stopwords

# Get the set of English stop words
stop_words = set(stopwords.words('english'))

# Define a function to remove stop words from tokenized words
def remove_stop_words(tokens):
    return [word for word in tokens if word not in stop_words]

# Apply the function to the last column and store the result in a new column
songs420['Stop Words'] = songs420.iloc[:, -1].apply(remove_stop_words)

```

```python
def flatten_if_nested(token_list):
    # If it's a list with exactly one item and that item is also a list, flatten it
    if isinstance(token_list, list) and len(token_list) == 1 and isinstance(token_list[0], list):
        return token_list[0]
    return token_list  # leave it as is if already flat

songs420['Tokenized'] = songs420['Tokenized'].apply(flatten_if_nested)
```

```python
brackets = {'[', ']'}
songs420['Tokenized'] = songs420['Tokenized'].apply(lambda tokens: [t for t in tokens if t not in brackets])
```

```python
from itertools import chain

def flatten_and_clean(tokens):
    # Flatten if nested (e.g., list of lists), and remove brackets
    flat = list(chain.from_iterable(tokens)) if tokens and isinstance(tokens[0], list) else tokens
    return [t for t in flat if isinstance(t, str) and t not in brackets]

songs420['Tokenized'] = songs420['Tokenized'].apply(flatten_and_clean)
```

```python
brackets = {'[', ']'}

def remove_brackets_safely(tokens):
    if isinstance(tokens, list):
        return [t for t in tokens if isinstance(t, str) and t not in brackets]
    return []

songs420['Tokenized'] = songs420['Tokenized'].apply(remove_brackets_safely)
```

```python
# Define a function to remove stop words from tokenized words
def remove_stop_words(tokens):
    return [word for word in tokens if word.lower() not in stop_words]

songs420['No Stop Words'] = songs420['Tokenized'].apply(remove_stop_words)
```

```python
songs420['No Stop Words'].head(7)
```

```python
import string

def remove_punctuation(tokens):
    return [word for word in tokens if word not in string.punctuation]

songs420['No Punctuation'] = songs420['Tokenized'].apply(remove_punctuation)
```

```python
from nltk.stem import PorterStemmer


stemmer = PorterStemmer()

def stem_tokens(tokens):
    return [stemmer.stem(token) for token in tokens]

songs420['Stemmed'] = songs420['No Punctuation'].apply(stem_tokens)
```
