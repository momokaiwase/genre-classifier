```python
import pandas as pd
```

```python
bottom_half = pd.read_csv("tiffanysongs422.xlsx - tiffanysongs422.xls.csv")
top_half = pd.read_csv("ricardosongs422.xlsx - ricardosongs422.xls.csv")
```

```python
df_combined = pd.concat([top_half, bottom_half], ignore_index=True)
```

```python
songs422 = pd.read_csv("songs422.csv")
```

```python
df_combined
```

```python
df_combined.shape
```

```python
songs422_labeled = pd.merge(songs422, df_combined[['Artist', 'Song Title', 'Genre']], on=['Artist', 'Song Title'], how='left')
```

```python
songs422_labeled
```

```python
missing_genre_count = songs422_labeled['Genre'].isnull().sum()
```

```python
print(missing_genre_count)
```

```python
songs422_labeled = songs422_labeled.dropna()
```

```python
songs422_labeled
```

```python
from nltk.tokenize import word_tokenize

# Make sure to download the NLTK tokenizer resources
import nltk
nltk.download('punkt')

# Tokenizing words in the 'Song Title' column
songs422_labeled['Song Title'] = songs422_labeled['Song Title'].apply(word_tokenize)
```

```python
df_combined.columns = ['Artist', 'Song Title', 'Genre']
```
